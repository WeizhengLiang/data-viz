```python
# Import statements
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# import data

# 1. First-born children by age of mother
# source: https://www.cbs.nl/en-gb/news/2019/19/average-age-of-first-time-mothers-up-to-29-9-years
firstBorn_age = pd.read_csv('/Users/weizhengliang/Desktop/datavis_hw/hw2/First-born children by age of mother.csv')


# 2. Mean age of mothers at birth of first child
# source: https://www.cbs.nl/en-gb/news/2019/19/average-age-of-first-time-mothers-up-to-29-9-years
mean_age = pd.read_csv('/Users/weizhengliang/Desktop/datavis_hw/hw2/mean_age.csv')

# 3. Percent distribution of women in the civilian labor force, 25 to 64 years of age, by educational attainment
# source: https://www.bls.gov/opub/ted/2017/a-look-at-womens-education-and-earnings-since-the-1970s.htm
women_edu = pd.read_csv('/Users/weizhengliang/Desktop/datavis_hw/hw2/women_education.csv')

# compare the 1st and 3rd data
born_edu = pd.read_csv('/Users/weizhengliang/Desktop/datavis_hw/hw2/born_edu.csv')
```

```python
# 

mean_age

plt.figure(figsize = (16,8), dpi = 300)
plt.plot('Year','Mean age at birth of first child (yrs)', data = mean_age, label = 'mean age')

plt.xticks(np.arange(1958, 2020, 10))
plt.yticks(np.arange(23, 33, 1))

plt.title('Mean age of mothers at birth of first child', fontsize = 15)
plt.ylabel('age(yrs)', fontsize = 15)



plt.grid()

plt.legend()

plt.show()
```

```python
# 'First-born children by age of mother' data
firstBorn_age

# Create figure
plt.figure(figsize = (10,5), dpi = 300)

# Create bar plot
pos = np.arange(6)
width = 0.2
plt.bar(pos-width*1.05, firstBorn_age['2008 (x 1,000)'], width, label = '2008 (x 1,000)')
plt.bar(pos, firstBorn_age['2013 (x 1,000)'], width, label = '2013 (x 1,000)')
plt.bar(pos+width*1.05, firstBorn_age['2018 (x 1,000)'], width, label = '2018 (x 1,000)')

# Specify ticks
plt.xticks(pos, rotation = 0)
plt.yticks(np.arange(0, 31, 10))

# Get current Axes for setting tick labels and horizontal grid
ax = plt.gca()

# Set tick labels
ax.set_xticklabels(firstBorn_age['Age'])

# Add major horizontal grid with dashed lines
ax.yaxis.grid(which = 'major', linestyle = '--')

# Add title
plt.title('First-born children by age of mother')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
women_edu

# Create figure
plt.figure(figsize = (10,6), dpi = 300)

# Create stacked area chart
label = women_edu.columns[1:]
plt.stackplot('Year', 'Less than a high school diploma', '4 years of high school', 
              '1 to 3 years of college', "Bachelor's degree and higher", data = women_edu, labels = label)
# Specify ticks
#plt.xticks(pos, rotation = 0)
plt.yticks(np.arange(0, 100, 25))
#plt.xticks(np.arange(1970, 2016, 9))

# Add legend
plt.legend()

# Add labels and title
plt.xlabel('women population in percentage', fontsize=15)
# Show plot
plt.show()
```

```python
plt.figure(figsize = (16,8), dpi = 300)

pos = np.arange(47)
width = .35 

born_edu["Bachelor's degree and higher"].plot(kind='bar', width = width)
ax = plt.gca()
plt.legend()
born_edu['Mean age at birth of first child (yrs)'].plot(secondary_y=True, color = 'r')


plt.xlim([-width, len(born_edu['Mean age at birth of first child (yrs)'])-width])


ax.set_xticklabels(born_edu['Year'], rotation = 70)


plt.legend(bbox_to_anchor=(0.251, 0.94))

plt.show()
```
